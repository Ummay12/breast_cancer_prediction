```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r"/content/breast-cancer (1).csv")
dataset
```

```python
del dataset['id']
```

```python
dataset.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
dataset['diagnosis'] = encoder.fit_transform(dataset['diagnosis'])
dataset['diagnosis'].unique()
```

```python
dataset
```

```python
dataset['diagnosis'].value_counts()
```

```python
data=dataset[[col for col in dataset.columns if col!='diagnosis']+['diagnosis']]
data
```

```python
x = data.iloc[:,:-1].values
y = data.iloc[:, -1].values
```

```python
import pandas as pd

major = data[data['diagnosis'] == 0]
minor = data[data['diagnosis'] == 1]
# Upsample the minority class
minor_upsampled = minor.sample(len(major), replace=True, random_state=42)
# Combine both
balanced_data = pd.concat([major, minor_upsampled])

# Shuffle the rows (optional but recommended)
balanced_data = balanced_data.sample(frac=1, random_state=42).reset_index(drop=True)

# Split into X and y
X_resampled = balanced_data.drop('diagnosis', axis=1)
y_resampled = balanced_data['diagnosis']
```

```python
from collections import Counter
print(Counter(y_resampled))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_resampled, test_size=0.2, random_state=15)
```

```python
ann = tf.keras.models.Sequential()
ann
```

```python
for _ in range(9):  # Loop to add 9 layers with 8 units each
    ann.add(tf.keras.layers.Dense(units=8, activation='relu'))

# Output layer with a single neuron and sigmoid activation
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

ann.fit(X_train, y_train, batch_size=30, epochs=50)
```

```python
ann_pred = ann.predict(X_test)
ann_pred
```

```python
New_ann_pred=[]
for x in ann_pred:
    if x >= 0.5:
        New_ann_pred.append(1)
    else:
        New_ann_pred.append(0)
print(New_ann_pred)
```

```python
from sklearn.metrics import accuracy_score
ANN_score = accuracy_score(y_test, New_ann_pred)
ANN_score*100
```
